## Conversión a datos mensuales:

Como el objetivo de nuestro proyecto es determinar el **consumo mensual** de determinados productos por unidad de
consumo, órgano gestor y provincia; lo primero que vamos a realizar es la agrupación de nuestros datos diarios a datos mensuales. Para ello realizaremos la suma de la cantidad de salida para todos aquellos datos que coincidan en artículo, centro de consumo, provincia, ubicación virtual, año y mes. 

Por lo tanto, lo primero que tenemos que realizar es la eliminación en los conjuntos train y test de aquellas variables que no nos son de utilidad, como son en este caso el identificador, la fecha en formato string, la fecha de caducidad de existencias y el día.

### Train:

```python
del Data_train['ID'], Data_train['FECHA'], Data_train['FECHACADUCIDADEXISTENCIAS'], Data_train['Dia']
```

```python
ImporteSalidaSuma = (Data_train.groupby(['ARTICULO', 'CENTROCONSUMO', 'DESC_PROVINCIA', 'ID_ORGANOGESTOR', 'Año', 'Mes']) # Agrupar
          .IMPORTESALIDA    # Quedarse con la columna IMPORTESALIDA
          .apply(sum)       # Calcular su suma
          .reset_index())   # Deshacer índice jerárquico

CantidadSalidaSuma = (Data_train.groupby(['ARTICULO', 'CENTROCONSUMO', 'DESC_PROVINCIA','ID_ORGANOGESTOR', 'Año', 'Mes']) # Agrupar
          .CANTIDADSALIDA    # Quedarse con la columna CANTIDADSALIDA
          .apply(sum)       # Calcular su suma
          .reset_index())   # Deshacer índice jerárquico

lista_cols = ['ARTICULO', 'CENTROCONSUMO', 'DESC_PROVINCIA', 'Año', 'Mes']
Data_train_mensual = pd.merge(ImporteSalidaSuma, CantidadSalidaSuma,  how='inner', on= lista_cols)
Data_train_mensual = Data_train_mensual.rename(columns={"ID_ORGANOGESTOR_x": "ID_ORGANOGESTOR"})
del Data_train_mensual['ID_ORGANOGESTOR_y']
print('Training dataset has:', Data_train_mensual.shape[0], 'rows and ', Data_train_mensual.shape[1], 'columns')
```

# Limpieza

En este apartado vamos a proceder a limpiar nuestro conjunto de datos para poder realizar el entrenamiento de nuestro modelo predictivo. Para ello, vamos a indigar qué variables tienen valores perdidos y en el caso de tenerlos, de qué manera vamos a proceder a imputarlos.

```python
Master.isnull().sum()
```

**Nota:** Vemos que las únicas variables que tienen valores perdidos o NA's son Temp_Max y Temp_Max.             

## Imputación de valores perdidos

### Temp_Max y Temp_Min:

Para corregir los valores NA's de estas variables, vamos a proceder a imputarlos de la siguiente manera:

- Si existen valores en el histórico que tengan el mismo centro de consumo y mes que el valor a imputar, se calculará la media de estos valores para las variables Temp_Max y Temp_Min; y se sustituirán dichos valores vacíos por éstas.
- En caso contrario, nos quedaremos con la media de Temp_Max y Temp_Min global para el mes coincidente con el del valor a imputar.

```python
index_empty = Master.loc[Master['Temp_Max'].isnull()].index.tolist()
for i in index_empty:
    A = (Master["CENTROCONSUMO"] == Master["CENTROCONSUMO"][i])
    B = (Master["Mes"] == Master["Mes"][i])
    average_Temp_Max = Master.loc[A & B]['Temp_Max'].mean()
    average_Temp_Max_Tot = Master.loc[B]['Temp_Max'].mean()
    average_Temp_Min = Master.loc[A & B]['Temp_Min'].mean()
    average_Temp_Min_Tot = Master.loc[B]['Temp_Max'].mean()
    if np.isnan(average_Temp_Max):
        Master['Temp_Max'][i] = average_Temp_Max_Tot
        Master['Temp_Min'][i] = average_Temp_Min_Tot
    else:
        Master['Temp_Max'][i] = average_Temp_Max
        Master['Temp_Min'][i] = average_Temp_Min
```

```python
Master.isnull().sum()
```

**Nota:** Podemos observar que ya no existen valores nulos en nuestro dataset.

# Modelo predictivo:  Red Neuronal Regresora

Una Red Neuronal Artificial o RNA está formada por una pila, normalmente ordenadas en forma de capas, de nodos/unidades/neuronas de cálculo que permiten pasar y transformar un conjunto de datos de entrada (dados en forma de vector) en una salida deseada. Estas unidades de cálculo elementales suelen, al menos en su forma más simple, tener una estructura simple que diferencia una etapa de agregación con pesos y una etapa de activación, lo que se conoce como **Perceptrón**. A partir de estas unidades simples se puede construir una Red de Perceptrones por medio de un apilamiento de capas, dando lugar a lo que se conoce como Perceptrón Multicapa **(MLP, por sus siglas en inglés)**. El siguiente diagrama muestra el esquema o estructura básifca de un MLP:

![Vanilla neural network](http://drive.google.com/uc?export=view&id=1e3uw9SYoOPCeW10_zB86FVEsR4_UsyXt)

Para evaluar la red mientras ajustamos los hiperparámetros (como el número de epochs o el Batch size a usar), vamos a poner en práctica una **K-validación cruzada**, que consiste en dividir los datos disponibles en $K$ particiones, después instanciar $K$ modelos exactos, y entonces entrenar cada uno de los modelos usando $K-1$ de las particiones anteriores y evaluar sobre la restante. La métrica de evaluación será la media de las $K$ métricas obtenidas sobre las validaciones de cada modelo.

A continuación, pasaremos a realizar el entrenamiento de nuestro modelo predictivo a través de una red neuronal regresora. Para el correcto desarrollo de éste, se ha creado previamente una librería que realiza todo los steps necesarios, entre los que podemos destacar:

- Normalización del dataset. 

- Realización del Split entre los conjuntos train y test.

- Definición de 6 tipos distintos de redes neuronales. Éstos son:
   1. Three hidden layers (relu: 64, 128, 64) and rmsprop optimizer
   2. Two hidden layers (relu: 16) and rmsprop optimizer
   3. One hidden layers (relu: 512) and rmsprop optimizer
   4. Three hidden layers (relu: 64) and rmsprop optimizer
   5. One hidden layers (sigmoid: 16) and rmsprop optimizer
   6. Four hidden layers (relu: 64) and rmsprop optimizer


- Entrenamiento de cada K-validación cruzada para cada red neuronal.

- Creación de una representación gráfica del mean absolute error (MAE) de cada epoch para cada red neuronal.

- Obtención de la media de las K métricas obtenidas sobre las validaciones para cada red neuronal.

- Evaluación del conjunto test para cada red neuronal.


Los distintos **outputs** creados por el modelo, serán guardados en las siguientes **carpetas:**

1. **Standard Scaler**: 'temp/scaler.pickle' and './temp/std_scaler.bin'
2. **Array y csv de las variables X del conjunto de entrenamiento**: './temp/train_X.pickle' y './temp/train_X.csv'
3. **Array y csv de la variable y del conjunto de entrenamiento**: './temp/train_Y.pickle' y './temp/train_Y.csv'
4. **Array y csv de las variables X del conjunto test**: './temp/test_X.pickle' y './temp/test_X.csv'
5. **Array y csv de la variable y del conjunto de test**: './temp/test_Y.pickle' y './temp/test_Y.csv'
6. **Modelo NNT entrenado**: './models/Model'+ i + '.h5' y 'models/NNTModel' + i + '.pickle' *(siendo **i** el número del modelo)*.
7. **Plot del K Average Mean Absolute Error por epoch**: './visuals/MAE Model'
8. **Mean Absolute Error del conjuto test**: 'temp/NNT_error' + i + '.pickle' *(siendo **i** el número del modelo)*.
